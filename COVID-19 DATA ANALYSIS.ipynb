```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True,axis=1)
```

```python
covid_df["Date"]=pd.to_datetime(covid_df['Date'],format="%Y-%m-%d")
```

###  total number of active cases


### active cases = total number of confirmed cases -(death + cured)

```python
covid_df["active_cases"]=covid_df["Confirmed"]-covid_df["Cured"]+covid_df["Deaths"]
```

```python
covid_df.tail()
```

```python
covid_df
```

### pivot table

```python
statewise = pd.pivot_table(covid_df,values = ["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise
```

### Recovery Rate

```python
statewise["Recocery rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

### Mortality Rate

```python
statewise["Mortality rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

## Sorting

```python
statewise = statewise.sort_values(by = "Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

### Top 10 active cases states

```python
plt.title("Top 10 states with most active cases in India",size=25)
```

```python
axs = sns.barplot(data = top_active_cases_states.iloc[:10],y = "active_cases",x = "State/UnionTerritory",linewidth = 2,edgecolor = 'red')
```

### TOP 10 STATES WITH HIGHEST STATES 

```python
top10_deaths_states = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by='Deaths',ascending=False).reset_index()
fig = plt.figure(figsize =(18,5))
plt.title("TOP 10 STATES WITH HIGHEST DEATHS",size=25)
ax =sns.barplot(data=top10_deaths_states.iloc[:12],y = "Deaths",x = "State/UnionTerritory",linewidth = 2,edgecolor="black")
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

### remove rows where state = India

```python
vaccine = vaccine_df[vaccine_df.State!="India"]
```

```python
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
```

```python
vaccine
```

### Most Vaccinated State

```python
most_vaccinated_state=vaccine.groupby("State")["Total"].sum().to_frame("Total")
```

```python
most_vaccinated_state = most_vaccinated_state.sort_values("Total",ascending=False)[:5]
```

```python
fig = plt.figure(figsize=(18,5))
plt.title("TOP 5 VACCINATED STATES IN INDIA",size=20)
ax = sns.barplot(data=most_vaccinated_state.iloc[:10],y = most_vaccinated_state.Total,x = most_vaccinated_state.index,linewidth=2,edgecolor="black")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

### Least Vaccinated State

```python
least_vaccinated_state=vaccine.groupby("State")["Total"].sum().to_frame("Total")
```

```python
least_vaccinated_state=least_vaccinated_state.sort_values("Total", ascending=True)[:5]
```

```python
least_vaccinated_state
```

```python
fig=plt.figure(figsize=(18,5))
plt.title("Top Least Vaccinated State",size=20)
ax = sns.barplot(data=least_vaccinated_state.iloc[:10],y=least_vaccinated_state.Total,x =least_vaccinated_state.index,linewidth=2,edgecolor="black")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
